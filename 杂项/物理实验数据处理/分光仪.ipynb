```python
import pandas as pd
import numpy as np

def deg_to_float(df: pd.DataFrame) -> pd.DataFrame:
    df = df.apply(lambda x: x.astype(int) + (x - x.astype(int)) * 100 / 60)
    df = df.apply(np.radians)
    return df
def float_to_deg(df: pd.DataFrame) -> pd.DataFrame:
    df = df.apply(np.degrees)
    df = df.apply(lambda x: x.astype(int) + (x - x.astype(int)) * 60 / 100)
    return df

df = pd.DataFrame([
    [59.32, 239.58, 298.31, 119.33, 0, 0, 0],
    [59.31, 239.48, 299.31, 119.34, 0, 0, 0],
    [59.33, 239.27, 299.31, 119.34, 0, 0, 0],
    [59.29, 239.28, 299.30, 119.29, 0, 0, 0],
    [59.32, 239.30, 299.32, 119.32, 0, 0, 0],
    [59.34, 239.31, 299.31, 119.33, 0, 0, 0]
    ],
                  columns=['AB1','AB2','AC1','AC2','phi1','phi2','phi'])

df = deg_to_float(df)

df['phi1'] = df['AB1'] - df['AC1'] + np.radians(360)
df['phi2'] = df['AB2'] - df['AC2']
df['phi'] = 1/2 * (df['phi1'] + df['phi2'])

df = float_to_deg(df)

df['phi'] = round(df['phi'],2)
df
```
