### Matplotlib基础绘图操作


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 ## 简单绘图


# % 生成数据
x = np.arange(0, 10, 0.1) # 从0到10，步长为0.1
y = np.sin(x) # 计算x的正弦值

# % 绘制图形
plt.plot(x, y) # 绘制x和y的关系
plt.title('Sin Wave') # 设置标题
plt.show() # 显示图形
```

```python
 ## 图像标签

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label='sin(x)') # 绘制sin(x)并设置标签
plt.plot(x,y2, linestyle = "--", label='cos(x)') # 虚线绘制cos(x)并设置标签

plt.title('Sin and Cos Waves') # 设置标题那那你再背你再背你再背
plt.xlabel('x') # 设置x轴标签
plt.ylabel('y') # 设置y轴标签
plt.legend() # 显示图例

plt.show() # 显示图形
```
