# XPath 解析实战

本 Notebook 由 `XPath解析.py` 转换而来，按逻辑拆分为语法说明与抓取示例。

## 1. 准备：请求并解析 HTML

```python
import lxml.html
import requests

# 获取 HTML
response = requests.get('http://books.toscrape.com/')
html = response.text

# 基于 HTML 内容创建选择器对象
selector = lxml.html.fromstring(html)
```

## 2. XPath 常用语法

### 路径规则
- `/html`：从根节点开始
- `//xxx`：从任意位置开始匹配 `xxx` 标签

### 基础找法
- `//标签名[@属性名='属性值']/...`：定位标签
- `//标签名[@属性名='属性值']/.../text()`：获取文本
- `//标签名[@属性名='属性值']/.../@属性名`：获取属性值

### 进阶找法
- `.../标签名[contains(@属性名,'关键字')]/...`：属性值包含某字符串
- `.../标签名[starts-with(@属性名,'开头部分')]/...`：属性值以某字符串开头

> ### 编写思路
> 从目标内容倒推，优先找“有标志性属性”的父标签，再逐层缩小范围。
>  

## 3. 示例：提取书名与链接

```python
# 获取文本
get_text = '//h3/a/text()'

# 获取属性
get_attr = '//h3/a/@href'

# 执行 XPath 并获取信息
title = selector.xpath(get_text)
attr = selector.xpath(get_attr)
print(title, attr)
```

### 先抓大后抓小

先抓到整块结构（每本书的 `article`），再在局部节点中提取字段，通常更稳健。

```python
# 先抓到所有书籍块
get_article = '//li/article[@class="product_pod"]'
article = selector.xpath(get_article)

# 再从每个书籍块中抓书名（这里演示第 1 本）
get_title = 'h3//a/text()'
title = article[0].xpath(get_title)

article[0], title
```

### string(.)方法

```python
article = selector.xpath(get_article)

# 直接获取整块的文字信息
info = article[0].xpath('string(.)')
print(info) #/ 包括大量换行和空格
```
